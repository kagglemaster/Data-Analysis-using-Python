```python
import pandas as pd
```

```python
data = pd.read_excel("Product_Info.xlsx", index_col="ProductID")
data
```

```python
data["QuantityInStock"].isnull().sum()
```

```python
data["QuantityInStock"].fillna(data["QuantityInStock"].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
def availability(quantity):
    if quantity < 60:
        return "Reorder Needed"
    else:
        return "Available"
```

```python
data["Availability"] = data["QuantityInStock"].apply(availability)
```
