```python
customer_data.Phone_Number = customer_data.Phone_Number.str.replace("/","").str.replace("|","").str.replace("-","").str.replace("Na--","N/a")
customer_data.Phone_Number.astype("string")
customer_data.Phone_Number = customer_data.Phone_Number.apply(lambda x: x[:3] + "-" + x[3:6] + "-" +x[6:])
```
